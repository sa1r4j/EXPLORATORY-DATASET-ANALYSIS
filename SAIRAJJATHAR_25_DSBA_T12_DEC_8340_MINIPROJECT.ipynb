```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/MyDrive/INTERNSELITE/cleaned_Diwali_sales.csv')
```

```python
print("Before:", df.shape)
```

```python

df.head()
```

```python

df.tail()
```

```python
print("\nMissing values:\n", df.isnull().sum())
```

```python
df["Amount"]=df["Amount"].fillna(df["Amount"].median())
```

```python
df['age_group'] =df['age_group'].fillna(df['age_group'].mode()[0])
```

```python
print("\nMissing values:\n", df.isnull().sum())
```

```python
df.describe()
```

```python
df = df.drop(['User_ID', 'Cust_name','Unnamed: 0','Zone'], axis=1)
```

```python
df = df.drop_duplicates()
```

```python
print("After:", df.shape)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
gender_sales = df.groupby('Gender')['Amount'].sum().sort_values()

plt.figure(figsize=(6,4))
plt.ticklabel_format(style='plain', axis='y')
gender_sales.plot(kind='bar', color='cyan')
plt.title("Total Purchase Amount by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Amount Spent")
plt.show()
```

```python
age_sales = df.groupby('Age Group')['Amount'].sum().sort_values()

plt.figure(figsize=(8,4))
plt.ticklabel_format(style='plain', axis='y')
age_sales.plot(kind='bar', color='cyan')
plt.title("Purchase Amount by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Amount Spent")
plt.show()
```

```python
state_sales = df.groupby('State')['Amount'].sum().sort_values(ascending=True)

plt.figure(figsize=(10,4))
plt.ticklabel_format(style='plain', axis='y')
state_sales.plot(kind='bar', color='cyan')
plt.title("Total Sales by State")
plt.xlabel("State")
plt.ylabel("Total Amount")
plt.show()
```

```python
occ_sales = df.groupby('Occupation')['Amount'].sum().sort_values()

plt.figure(figsize=(8,4))
occ_sales.plot(kind='bar', color='cyan')

plt.ticklabel_format(style='plain', axis='y')
plt.title("Purchase Amount by Occupation")
plt.xlabel("Total Amount")
plt.ylabel("Occupation")
plt.show()
```

```python
cat_orders = df.groupby('Product_Category')['Orders'].sum().sort_values(ascending=True)

plt.figure(figsize=(9,6))
cat_orders.plot(kind='bar', color="cyan")
plt.title("Most Purchased Product Categories")
plt.ylabel("Number of Orders")
plt.xlabel("Category")
plt.show()
```

```python
marital_sales = df.groupby('Marital_Status')['Amount'].sum()

plt.figure(figsize=(5,4))
marital_sales.plot(kind='bar',color="cyan")
plt.ticklabel_format(style='plain', axis='y')
plt.title("Purchase by Marital Status")
plt.xlabel("Marital Status (0 = Single, 1 = Married)")
plt.ylabel("Total Amount Spent")
plt.show()

```

```python
city_sales = df.groupby("State")['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(8,4))
city_sales.plot(kind='bar', color='cyan')
plt.ticklabel_format(style="plain",axis="y")
plt.title("total sales by state")
plt.xlabel("state")
plt.ylabel("total amount")
plt.show()
```

```python
plt.figure(figsize=(7,4))
plt.hist(df["Amount"],bins=25,color="cyan")
plt.title("Distribution of purchase amount")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
df[df["Gender"]=="F"]['Amount'].hist(alpha=0.5)
df[df["Gender"]=="M"]['Amount'].hist(alpha=0.5)
plt.legend(["Female","Male"])
plt.xlabel("Amount")
plt.title("purchase amount by gender")
plt.show()
```
